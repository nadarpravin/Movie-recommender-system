We need a metric for rating the movie
We need to Calculate the score for every movie
Sort the scores and recommend the best rated movie to the users.
A simple way out to find the score is to go with the average ratings,
but using this won’t be fair enough since a movie with 8.7 average rating and only 8 votes cannot be considered better than the movie with 7.9 as as average rating but 40 votes,
let us use IMDB’s weighted rating (wr) which is given as :-

weighted rating=(v/(v+m)*R)+(v/(v+m)*C)

```python
C=df2['vote_average'].mean()
print(C)   # it shows mostly all movies are rated near 6 out of 10
```

The next step is to determine an appropriate value for m,
the minimum votes required to be listed in our list of good movies. We will use 90th percentile as our cutoff. In other words,
for a movie to feature in the list, it must have more votes than at least 90% of the movies in the list.

```python
m=df2['vote_count'].quantile(0.9)
print(m)
```

```python
list_movies=df2.copy().loc[df2['vote_count'] >=m]
list_movies.head(2)
```

```python
list_movies.shape
```

```python
#we find that 481 are the best movies with the quantile of 90%
```

```python
def weighted_rating(x,m=m,C=C):
    v=x['vote_count']  #v is the number of votes for the movie;
    
    R=x['vote_average']  #R is the average rating of the movie
    
    #calculation based on imdb formula (m=1838,c=6.09)
    return (v/(v+m)*R)+(v/(v+m)*C)
```

```python
#defining a new feature score and calculating its value with weighted rating
```

```python
list_movies['score']=list_movies.apply(weighted_rating,axis=1)
```

```python
list_movies['score'].head(2)
```

```python
#sorting movies based on weighted average
```

```python
list_movies=list_movies.sort_values('score',ascending=False)
```
